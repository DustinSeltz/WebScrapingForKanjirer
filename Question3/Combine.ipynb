Description: Combine data from Jisho and Wanikani.

Input: 'jisho_result.csv' 'wanikani_result.csv' 'genki.csv'

Output: 'combined.csv' 'combined_genki.csv'

```python
import pandas as pd
```

```python
jisho = pd.read_csv("jisho_result.csv")
wani = pd.read_csv("wanikani_result.csv")
```

```python
jisho.head()
```

```python
wani.head()
```

```python
jisho['wanikani_level'] = wani['level']
```

```python
jisho.head()
```

```python
jisho.to_csv('combined.csv')
```

```python
genki = pd.read_csv('genki.csv')
```

```python
newdf = jisho.merge(genki, on = 'kanji', how = 'left')
```

```python
newdf.to_csv('combined_genki.csv')
```
