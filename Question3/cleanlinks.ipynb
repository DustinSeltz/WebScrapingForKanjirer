Input: 'combined_genki_lessons.csv'
Output: 'cleaned_link.csv'

```python
import pandas as pd
import ast
```

```python
file = pd.read_csv('combined_genki_lessons.csv')
```

```python
file
```

```python
d = ast.literal_eval(file['parts'][0])
```

```python
newcol = [None] * len(file['parts'])
for i in range(0,len(file['parts'])):
    newcol[i] = list(ast.literal_eval(file['parts'][i]).keys())
```

```python
len(newcol)
```

```python
file['parts'] = newcol
```

```python
file
```

```python
newcol = [None] * len(file['on_readings'])
for i in range(0,len(file['on_readings'])):
    try:
        newcol[i] = list(ast.literal_eval(file['on_readings'][i]).keys())
    except:
        continue
len(newcol)
```

```python
file['on_readings']  = newcol
```

```python
file
```

```python
newcol = [None] * len(file['kun_readings'])
for i in range(0,len(file['kun_readings'])):
    try:
        newcol[i] = list(ast.literal_eval(file['kun_readings'][i]).keys())
    except:
        continue
len(newcol)
```

```python
file['kun_readings'] = newcol
```

```python
file
```

```python
file.to_csv('cleaned_link.csv')
```
