```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = "http://genki.japantimes.co.jp/self/genki-kanji-list-linked-to-wwkanji"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
things = soup.find_all("td")
lesson = []
kanji = []
curr = 0
for i in range(1,len(things)):
    l = re.search(r'Lesson (.*):', things[i].getText())
    if not l == None:
        curr = l[1]
    else:
        if not things[i].getText() == '\xa0':
            kanji.append(things[i].getText())
            lesson.append(curr)
```

```python
lesson
```

```python
df = pd.DataFrame({'kanji':kanji, 'Genki_Level': lesson})
```
